```python
import numpy as np
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt

# Data points
x = [2, 3, 4, 5, 6, 7]
y = [12, 12, 32, 16, 22, 25]

# Linear spline interpolation
linear_interp = interp1d(x, y, kind='linear')

# Quadratic spline interpolation
quadratic_interp = interp1d(x, y, kind='quadratic')

# Generating points for plotting
x_interp = np.linspace(2, 7, 100)
y_linear = linear_interp(x_interp)
y_quadratic = quadratic_interp(x_interp)

# Plot the original data points and the interpolation
plt.figure(figsize=(10, 6))
plt.scatter(x, y, label='Data Points')
plt.plot(x_interp, y_linear, label='Linear Spline', linestyle='--')
plt.plot(x_interp, y_quadratic, label='Quadratic Spline', linestyle='-.')
plt.legend()
plt.title('Linear and Quadratic Spline Interpolation')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.show()
```
